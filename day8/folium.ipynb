3. 주변지역을 묶음 : from folium.plugins import Markercluster, MarkerCluster().add_to(map)

```python
# 서울시 반려견 놀이터 지도 시각화

import pandas as pd

df = pd.read_csv('day8_data\서울시반려견공원.csv')
df
```

```python
from folium.plugins import MarkerCluster

# folium 객체 생성
map = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=10)
marker = MarkerCluster().add_to(map)

for i in df.index:
    folium.Marker(location=[df.loc[i, 'Latitude'], df.loc[i, 'Longitude']], popup=df.loc[i, '공원명'], icon=folium.Icon(color='green', icon='dog', prefix='fa')).add_to(marker)

map.save('map.html')
```
