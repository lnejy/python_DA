```python
import pandas as pd
```

```python
score1 = pd.read_csv('day3_data\중간기말성적.csv', encoding='cp949')
score1.head()
```

```python
score1[score1['분반'] == 'A']
```

```python
# 성별이 '여'인 데이터 추출

score1[score1['성별'] == '여']
```

```python
# 중간고사, 기말고사 모두 90 이상인 데이터 추출

score1[(score1['중간고사'] >= 90) & (score1['기말고사'] >= 90)]
```

```python
# 중간고사, 기말고사 중 하나가 90 이상인 데이터 추출

score1[(score1['중간고사'] >= 90) | (score1['기말고사'] >= 90)]
```

* 방법 2 : 변수.query('조건식')

```python
# 분반이 'B' 분반인 데이터 추출

score1.query('분반=="A"')
```

```python
# '남' 데이터이거나 '중간고사'가 100인 데이터 추출, or 연산

score1.query('성별=="남" | 중간고사 == 100')
```

```python
# 분반이 'A' 데이터 중에서 '중간고사', '기말고사' 80 이상인 데이터

score1.query('분반=="A" & 중간고사>=80 & 기말고사>=80')
```

```python
score1.query('분반=="A" & 중간고사>=80 & 기말고사>=80')[['학번', '성별']]
```

### 데이터 정렬
* 변수명.sort_values('열이름', ascending=True/False)

```python
# 성별 기준 오름차순, 기말고사 기준 내림차순 정렬

score1.sort_values(['성별', '기말고사'],ascending=[True,False])
```

### 통계 함수
* 열이름.통계함수()

```python
# 중간고사, 기말고사 평균

meanData = score1[['중간고사', '기말고사']].mean()
type(meanData)
meanData
```

### 그룹화
* 변수명.groupby('그룹열일이름')[계산할 열이름].통계함수
* 그룹화할 열이 두개 이상을 경우에는 대괄호 사용

```python
# 분반을 기준으로 '기말고사' 평균 계산

score1.groupby('분반')['기말고사'].mean()
```

```python
# '분반', '성별' 깆준으로 중간고사, 기말고사 평균 계싼

score1.groupby(['분반', '성별'])[['중간고사', '기말고사']].mean()
```

### 1. score1 dataframe에서 기말고사 90 이상이거나 중간고사 90 이상인 데이터를 성별을 기준으로 기말고사와 중간고사의 평균 계산

```python
t1 = score1.query('중간고사>=90 | 기말고사>=90').groupby('성별')[['중간고사', '기말고사']].mean()
t1
```

### 1. score1 dataframe에서 기말고사 90 이상이거나 중간고사 90 이상인 데이터를 성별을 기준으로 인원수 계산

```python
t2 = score1.query('중간고사>=90 | 기말고사>=90').groupby('성별')[['중간고사']].count()
t2.rename(columns={'중간고사':'인원수'}, inplace=True)
t2
```

```python
t3 = pd.merge(t1,t2,on='성별')
t3
```
