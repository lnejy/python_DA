```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
```

```python
data = {'이름':['AAA', 'BBB', 'CCC', 'DDD'],
        '점수':[60, 70, 80, 90]}

df = pd.DataFrame(data)

# 판다스 plot을 이용한 차트 작성
# 데이터프레임.plot()
df.plot(kind='bar', x='이름', rot=0, ylim=[0,100])

for i in list(df.index):
    
    rate = f'{(df.loc[i]['점수']/df['점수'].sum())*100:.1f}%'
    plt.text(x=i-0.1, y=df.loc[i]['점수']+1, s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.2, y=df.loc[i]['점수']/2, s=rate)

meanValue = df['점수'].mean()
plt.axhline(y=meanValue, color='green')
plt.text(x=0, y=meanValue+1, s=f'평균:{meanValue:.1f}')

plt.xlabel('학생 이름')
plt.ylabel('점수')
plt.title('학생별 시험 점수')

plt.savefig('graph01.png', dpi=400)
print('파일 저장 완료')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')

# subplot 작성, matplotlib 이용
data = {'year':[2010, 2011, 2012, 2013, 2014],
        'sales':[400, 800, 1200, 1600, 2000],
        'expenses':[300, 400, 500, 600, 700]}

df2 = pd.DataFrame(data)

# subplot 생성
fig, axes = plt.subplots(2, 2, figsize=(10,9))

# 선 그래프
axes[0,0].plot(df2['year'], df2['sales'])
axes[0,0].set_title('sales', fontsize=12)
axes[0,0].set_xlabel('year', fontsize=9)

# 막대 그래프
axes[0,1].bar(df2['year'], df2['expenses'])
axes[0,1].set_title('expenses')

# 산점도 그래프
axes[1,0].scatter(df2['sales'], df2['expenses'])
axes[1,0].set_title('sales vs expenses')

# 히스토그램
axes[1,1].hist(df2['sales'], bins=4)
axes[1,1].set_title('sales distribution')


# 그래프 간 간격 조정
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='Malgun Gothic')

# subplot 작성, seabron 차드 작성
data = {'year':[2010, 2011, 2012, 2013, 2014],
        'sales':[400, 800, 1200, 1600, 2000],
        'expenses':[300, 400, 500, 600, 700]}

df3 = pd.DataFrame(data)

# subplot 생성
fig, axes = plt.subplots(2, 2, figsize=(10,9))

# 선 그래프
sns.lineplot(data=df3, x='year', y='sales', ax=axes[0,0])
axes[0,0].set_title('sales', fontsize=12)
axes[0,0].set_xlabel('year', fontsize=9)

# 막대 그래프
sns.barplot(data=df3, x='year', y='expenses', ax=axes[0,1], palette='Set1')

# 산점도 그래프
sns.scatterplot(data=df3, x='sales', y='expenses', ax=axes[1,0], palette='Set1')

# 히스토그램
sns.histplot(data=df3, x='sales', ax=axes[1,1], palette='Set1')

plt.show()
```
